## Regress√£o logistica 

nesse momento eu quero entender quanto meu modelo pode acertar se existe fraude ou n√£o no pipeline 

```python
x = df.drop('fraude', axis=1)    #fa√ßo um drop da coluna fraude para considerar apenas ela como teste e treino
y = df['fraude']
```

```python
from sklearn.model_selection import train_test_split ## defino o SEED e incluir os dados para rodar e separar no sklearn os dados teste 25% e treini 75%

SEED = 42

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = SEED)
```

```python
from sklearn.linear_model import LogisticRegression  #aqui apesar da acuracia esta aumenta minhas outras metricas es√£o muito 
from sklearn import metrics

lr = LogisticRegression(max_iter=1000, random_state=SEED)
lr.fit(x_train,y_train)
y_pred =lr.predict(x_test)


print("Acur√°cia:",metrics.accuracy_score(y_test, y_pred))
print("Precis√£o:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) 
print("F1:",metrics.f1_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt                      #constru√ß√£o da matriz de confusao para entender porque apesar da acuaracia estar alta a precisao esta baixa
from sklearn.metrics import Confusion_matrix
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay
%matplotlib inline


cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
y_pred_proba = lr.predict_proba(x_test)[::, 1]
fpr, tpr, _= metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.rcParams['figure.figsize'] = (12., 8.)
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
```
