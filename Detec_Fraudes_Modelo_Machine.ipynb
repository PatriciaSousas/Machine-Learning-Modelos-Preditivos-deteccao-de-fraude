## Regressão logistica 

nesse momento eu quero entender quanto meu modelo pode acertar se existe fraude ou não no pipeline 

```python
x = df.drop('fraude', axis=1)    #faço um drop da coluna fraude para considerar apenas ela como teste e treino
y = df['fraude']
```

```python
from sklearn.model_selection import train_test_split ## defino o SEED e incluir os dados para rodar e separar no sklearn os dados teste 25% e treini 75%

SEED = 42

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = SEED)
```

```python
from sklearn.linear_model import LogisticRegression  #aqui apesar da acuracia esta aumenta minhas outras metricas esão muito 
from sklearn import metrics

lr = LogisticRegression(max_iter=1000, random_state=SEED)
lr.fit(x_train,y_train)
y_pred =lr.predict(x_test)


print("Acurácia:",metrics.accuracy_score(y_test, y_pred))
print("Precisão:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) 
print("F1:",metrics.f1_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt                      #construção da matriz de confusao para entender porque apesar da acuaracia estar alta a precisao esta baixa
from sklearn.metrics import Confusion_matrix
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay
%matplotlib inline


cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
y_pred_proba = lr.predict_proba(x_test)[::, 1]    #construção da curva roc para validar meu teste
fpr, tpr, _= metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.rcParams['figure.figsize'] = (12., 8.)
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
```

## Balanceando de dados

#### Detectei usando a curva de Roc um desbaleaceamento dos dados classificados, acuracia está alta, porém as demais metrícas estão em desequilibrio no modelo, para fazer esse balanceamento 
#### vou usar o metodo Oversampling que esse caso vai ser útil e não terá problemas, pois aqui eu estou trabalhando com modelos de dados sinteticos, numa base real de dados eu teria que usar uma outra metodologia para balancear

```python
!pip install -U pandas-profiling
```

```python
from imblearn.over_sampling import SMOTE  #o SMOTE sintetiza novos exemplos para a classe minoritária
import numpy as np

SEED=42

smote = SMOTE(random_state=42)
```

```python
x = df.drop('fraude', axis=1)   #vou separar novamente os dados para que o SMOTE consiga classificar os dados de uma maneira correta
y = df['fraude']
```

```python
x_resampled, y_resampled = smote.fit_resample(x,y) #vou rodar o primeiro modelo
```

```python
df_balanced = pd.concat([y_resampled, x_resampled], axis=1) #abri um dataframe para incluir os dados do meu modelo 
df_balanced
```

```python
df_balanced.groupby('fraude').tempo.count() #Rodei o modelo e agora eu quero validar se os dados sem distribuidos de uma maneira igualitaria

fraude
0    101497
1    101497
Name: tempo, dtype: int64
```

```python
x = df_balanced.iloc[:, 1:].values
y = df_balanced.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split #vou incluir o SMOTE no meu modelo para entender como será os resultados

SEED = 42

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = SEED)

from sklearn.linear_model import LogisticRegression
from sklearn import metrics

lr = LogisticRegression(max_iter=1000, random_state=SEED)

lr.fit(x_train, y_train)
y_pred =lr.predict(x_test)

print("Acurácia:",metrics.accuracy_score(y_test, y_pred))
print("Precisão:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred)) 
print("F1:",metrics.f1_score(y_test, y_pred))
```
